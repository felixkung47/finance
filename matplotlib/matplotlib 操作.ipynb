#### 1.1單一曲線圖

資本資產定價模型(Capital Asset Pricing Model, CAPM), 美國William Sharpe提出。

$$ R_{i} = R_{F} + \beta(R_{M} - R_{F}) $$

* $R_{i}$是投資資產的預期收益率。
* $R_{F}$是無風險收益率。
* $R_{M}$是由所有可投資資產組合產生的收益率。也就是市場收益率(return on the market)。
* $\beta$是系統風險度量。衡量投資資產的收益率對市場收益率的敏感度。

證券市場線(Securities Market Line)反映單一股票或是投資組合的預期收益率與beta之間的關係。

```python
#涉某股票的beta在[0.5,2.0]之間，無風險利率為3%, 市場收益率是12%。請繪製該股票的證券市場線。
beta = np.linspace(0.5,2.0,100)
Rf = 0.03
Rm = 0.12
Ri = Rf + beta*(Rm-Rf)
```

```python
plt.figure(figsize=(9,6))
plt.plot(beta, Ri, 'r-', label = '證券市場線', lw=2.0)
plt.plot(1.0, Rf+1.0*(Rm-Rf), 'o', lw=2.5)
plt.axis('tight')
plt.xticks(fontsize=14)
plt.xlabel(u'beta值',fontsize=14)
plt.xlim(0.4, 2.1)
plt.yticks(fontsize=14)
plt.ylabel(u'單一股票收益率',fontsize=14, rotation=90)
plt.ylim(0.07, 0.22)
plt.title(u'資本資產定價模型', fontsize=14)
plt.annotate(u'beta等於1的收益', fontsize=14, xy=(1.0, 0.12), xytext=(0.8, 0.15), 
             arrowprops=dict(facecolor='b', shrink=0.05))
plt.legend(loc=0, fontsize=14)
plt.grid()
```

### 2.直方圖

#### 2.1單一樣本直方圖

```python
I = 1000
x_norm = npr.normal(loc=0.8, scale=1.5, size=I)
x_log = npr.lognormal(mean=0.5, sigma=1.0, size=I)
x_chi = npr.chisquare(df=4, size=I)
x_beta = npr.beta(a=2, b=4, size=I)
```

```python
plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
plt.hist(x_norm, label=u'常態分配的抽樣', bins=20, facecolor='y', edgecolor='k')
plt.xticks(fontsize=13)
plt.xlabel(u'樣本值', fontsize=13)
plt.yticks(fontsize=13)
plt.ylabel(u'頻率', fontsize=13, rotation=90)
plt.legend(loc=0,fontsize=13)
plt.grid()
plt.subplot(2,2,2)
plt.hist(x_log, label=u'對數常態分配的抽樣', bins=20, facecolor='r', edgecolor='k')
plt.xticks(fontsize=13)
plt.xlabel(u'樣本值', fontsize=13)
plt.yticks(fontsize=13)
plt.ylabel(u'頻率', fontsize=13, rotation=90)
plt.legend(loc=0,fontsize=13)
plt.grid()
plt.subplot(2,2,3)
plt.hist(x_chi, label=u'卡方分配的抽樣', bins=20, facecolor='b', edgecolor='k')
plt.xticks(fontsize=13)
plt.xlabel(u'樣本值', fontsize=13)
plt.yticks(fontsize=13)
plt.ylabel(u'頻率', fontsize=13, rotation=90)
plt.legend(loc=0,fontsize=13)
plt.grid()
plt.subplot(2,2,4)
plt.hist(x_beta, label=u'beta分配的抽樣', bins=20, facecolor='c', edgecolor='k')
plt.xticks(fontsize=13)
plt.xlabel(u'樣本值', fontsize=13)
plt.yticks(fontsize=13)
plt.ylabel(u'頻率', fontsize=13, rotation=90)
plt.legend(loc=0,fontsize=13)
plt.grid()
```
